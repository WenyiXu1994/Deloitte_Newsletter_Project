```python
# get the date of yesterday

def transfer_day():
    now_time = datetime.datetime.now()
    yes_time = now_time + datetime.timedelta(days=-1)
    if yes_time.day < 10:
        return '0'+str(yes_time.day)
    else:
        return str(yes_time.day)
transfer_day()
```

```python
# change the num month to text months
def num_to_month(num):
    months = {
        1 : "January",
        2 : "February",
        3 : "March",
        4 : "April",
        5 : "May",
        6 : "June",
        7 : "July",
        8 : "August",
        9 : "September",
        10 : "October",
        11 : "November",
        12 : "December"
    }
    return months.get(num)
```

```python
def techcrunch_():
    try:
        techcrunch_text = newtechcrunch()
        techcrunch_html = etree.HTML(techcrunch_text)
        techcrunch_text = etree.tostring(techcrunch_html).decode()
        xpath = '//div[@class = "post-block post-block--image post-block--unread"]/header[@class = "post-block__header"]/h2[@class = "post-block__title"]/a/@href'
        techcrunch_suburl = techcrunch_html.xpath(xpath)
        techcrunch_suburl
        techcrunch_news = []
        source = []
        urllink = []
        for url in techcrunch_suburl:
            techcrunch_news.append(url.split('/')[6].replace('-',' '))
            source.append(url.split('/')[2])
            urllink.append(url.replace('\'',''))
        techcrunch = {'news source':source,
                  'news content':techcrunch_news,
                      'link':urllink
                      }
        techcrunch = DataFrame(techcrunch)
        return techcrunch
    except:
        print("Error when getting web data")
    
techcrunch_()
```

```python
def forbes_():
    try:
        url = "https://www.forbes.com/innovation"
        r = requests.get(url)
        content = r.text
        soup = BeautifulSoup(content, 'lxml')
        
        urls = []
        titles = []
        a_s = soup.find_all(class_ = "csf-block")[0].find_all('a')
        for a in a_s:
            if 'headlink' in a.get('class'):
                urls.append(a.get('href'))
                titles.append(a.get_text())
        
        forbes = {
            'news source': ["https://www.forbes.com/innovation" for x in urls],
            'news content': titles,
            'link': urls
        }
        forbes =  DataFrame(forbes)
        return(forbes)
    except:
        print("Error when getting web data")
forbes_()
```

```python
def jike_():
    jike_text = newjike()
    jike_html = etree.HTML(jike_text)
    xpath1 = '//div[@class="article-time"]/text()'
    xpath2 = '//a[@data-event-category="article-list.title"]/@href'
    xpath3 = '//a/h3/text()'
    jike_time1 = jike_html.xpath(xpath1)
    jike_url1 = jike_html.xpath(xpath2)
    jike_news1 = jike_html.xpath(xpath3)
    
    jike_time = []
    jike_url = []
    jike_news = []
    jike_source = []
    for time in jike_time1:
        if len(time.split('/')) > 1:
            jike_time.append(time.split("/")[2])
        else:
            jike_time.append(time.split('/')[0])
    for url in jike_url1:
        jike_url.append("https://www.geekpark.net"+url)
    for news in jike_news1:
        jike_news.append(news)
    for i in range(0,len(jike_news)):
        jike_source.append("极客公园")
    jike_time = np.array(jike_time)
    jike_url = np.array(jike_url)
    jike_news = np.array(jike_news)
    jike_source = np.array(jike_source)
    
    jike_news = jike_news[jike_time == transfer_day()]
    jike_url = jike_url[jike_time == transfer_day()]
    jike_source = jike_source[jike_time == transfer_day()]
    
    jike = {
        "news source":jike_source,
        "news content":jike_news,
        "link":jike_url
    }
    jike = DataFrame(jike)
    return jike
jike_()
```

# Merge the news from all the websites above

```python
def collect_news():
    techcrunch1 = techcrunch_()
    businessinsider1 = businessinsider_()
    techmeme1 = techmeme_()
    venturebeat1 = venturebeat_()
    forbes1 = forbes_()
    zdnet1 = zdnet_()
    leifeng1 = leifeng_()
    jike1 = jike_()
    all_news = [techcrunch1,businessinsider1,techmeme1,venturebeat1,forbes1,zdnet1,leifeng1,jike1]
    all_news = pd.concat(all_news)
    all_news.index = range(len(all_news))
    
    time = datetime.datetime.now()
    today = str(time.year)+str(time.month)+str(time.day)
    all_news.to_csv(today+"news_content.csv",encoding ="utf_8_sig")
    return all_news
collect_news()
```
