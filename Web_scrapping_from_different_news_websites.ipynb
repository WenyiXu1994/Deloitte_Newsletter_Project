```python
now_time = datetime.datetime.now()
yes_time = now_time + datetime.timedelta(days=-1)
yes_time.day
def transfer_day():
    now_time = datetime.datetime.now()
    yes_time = now_time + datetime.timedelta(days=-1)
    if yes_time.day < 10:
        return '0'+str(yes_time.day)
    else:
        return str(yes_time.day)
transfer_day()
```

```python
def num_to_month(num):
    months = {
        1 : "January",
        2 : "February",
        3 : "March",
        4 : "April",
        5 : "May",
        6 : "June",
        7 : "July",
        8 : "August",
        9 : "September",
        10 : "October",
        11 : "November",
        12 : "December"
    }
    return months.get(num)
```

```python
def forbes_():
    try:
        forbes_text = newforbes()
        re_str = r'{"brandVoiceBrand":"","isInsights":false,"isMobile":false,"id":"(.*?)"cardClass":"card--'      #第一步筛选
        re = ree.compile(re_str)
        forbes_url = re.findall(forbes_text)
        forbes_url_ = []
        re_str2 = r'"url":(.*?),'     #第二个正则进一步筛选
        re2 = ree.compile(re_str2)
        for i in forbes_url:
            forbes_url_.append(re2.findall(i))
        forbes_url_ = reduce(operator.add,forbes_url_)
        today_time = datetime.datetime.now()
        yesterday_time = today_time + datetime.timedelta(days = -1)
        yesterday_time.day
        forbes_yes = []
        for url in forbes_url_:
            if url.split('/')[7] == transfer_day():
                forbes_yes.append(url)
        forbes_yes_news = []
        source = []
        for url in forbes_yes:
            url = url.replace(r'"','')
            source.append(url.split('/')[2])
            forbes_yes_news.append(url.split('/')[8].replace("-"," "))
        forbes = {
            '新闻源': source,
            '新闻内容': forbes_yes_news,
            '链接':forbes_yes
        }
        forbes =  DataFrame(forbes)
        return(forbes)
    except:
        print("爬取失败")
forbes_()
```
